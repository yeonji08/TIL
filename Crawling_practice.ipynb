```python
import requests
from bs4 import BeautifulSoup
```

```python
req = requests.get("https://m.blog.naver.com/luvpolo/221300832833") 
req.content
soup = BeautifulSoup(req.content, "html.parser")
```

```python
print(soup)
```

```python
span_list = soup.find("tbody").find_all("span")
```

```python
soup.find("tbody").find_all("span")[0].text
```

```python
for span in span_list:
    print(span.text)
```

```python
span_list2 = []
for span in span_list:
    span_list2.append(span.text)
print(span_list2)
```

```python
del span_list2[0]
```

```python
span_list2
```

```python
span_list3 = []
for span in span_list2:
    if len(span) > 3:
        span_list3.append(span)
print(span_list3)
        
```

```python
f = open('D:/temp/voca.txt', 'w')
for span in span_list3:
         f.write(span+"\n")
f.close()
```
